```python
df['last_review'] = df['last_review'].astype(str).fillna('No Review')


```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].median())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['price'], bins=50, kde=True)
plt.xlim(0, 500)
plt.title('Distribution of Listing Prices (Capped at $500)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Listings')
plt.show()
```

```python
sns.countplot(data=df, x='room_type', palette='Set2')
plt.title('Count of Room Types')
plt.xlabel('Room Type')
plt.ylabel('Number of Listings')
plt.show()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
df['last_review'].dt.to_period('M').value_counts().sort_index().plot(kind='line', title='Review Activity Over Time', marker='o')
```

```python
sns.barplot(data=df, x='neighbourhood_group', y='price', estimator='mean', palette='Set3')
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.show()

```

```python
availability_buckets = pd.cut(df['availability_365'], bins=[0, 90, 180, 270, 365],
                               labels=['090 days', '91180 days', '181270 days', '271365 days'])
availability_counts = availability_buckets.value_counts().sort_index()

plt.figure(figsize=(6, 6))
plt.pie(availability_counts, labels=availability_counts.index, autopct='%1.1f%%', startangle=120, colors=sns.color_palette('Set2'))
plt.title('Availability of Listings (Bucketed)')
plt.axis('equal')
plt.show()
```

```python
df['host_id'].value_counts().head(10).plot(kind='bar', title='Top 10 Hosts by Number of Listings', color='slateblue')

```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sample_df = df.sample(5000)  # for better performance in plotting
sns.scatterplot(data=sample_df, x='longitude', y='latitude', hue='neighbourhood_group', alpha=0.4, palette='muted')
plt.title('Geospatial Distribution of Airbnb Listings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
host_counts = df['host_id'].value_counts()
listing_type = host_counts.apply(lambda x: 'Multiple Listings' if x > 1 else 'Single Listing')
listing_type_counts = listing_type.value_counts()

plt.pie(listing_type_counts, labels=listing_type_counts.index, autopct='%1.1f%%', colors=['#8ecae6', '#219ebc'], startangle=60)
plt.title('Listings: Single vs Multiple per Host')
plt.axis('equal')
plt.show()
```

```python
sns.boxplot(data=df[df['price'] < df['price'].quantile(0.95)], x='room_type', y='price', palette='pastel')
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

```python
df[df['minimum_nights'] < 30]['minimum_nights'].plot(kind='hist', bins=30, color='skyblue', edgecolor='black', title='Minimum Nights Distribution (under 30 days)')
```

```python
sns.histplot(df['availability_365'], bins=30, kde=True, color='teal')
plt.title('Availability Distribution (Days per Year)')
plt.xlabel('Number of Days Available')
plt.ylabel('Count of Listings')
plt.show()
```

```python
top_neigh = df['neighbourhood'].value_counts().head(10)
sns.barplot(x=top_neigh.values, y=top_neigh.index, palette='flare')
plt.title('Top 10 Neighbourhoods by Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```

```python
top_hosts = df.groupby('host_name')['id'].count().sort_values(ascending=False).head(10)
sns.barplot(x=top_hosts.values, y=top_hosts.index, palette='coolwarm')
plt.title('Top 10 Hosts by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Host Name')
plt.show()
```

```python
sns.barplot(data=df, x='room_type', y='reviews_per_month', estimator='mean', palette='crest')
plt.title('Average Reviews per Month by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Reviews/Month')
plt.show()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'])
monthly_reviews = df['last_review'].dt.to_period('M').value_counts().sort_index()

monthly_reviews.plot(kind='line', marker='o', figsize=(10,4), color='purple')
plt.title('Monthly Number of Reviews')
plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.grid(True)
plt.show()


```

```python
price_bins = [0, 50, 100, 200, 300, 500, 1000]
df['price_range'] = pd.cut(df['price'], bins=price_bins)
df['price_range'].value_counts().sort_index().plot(kind='bar', color='teal', title='Listings by Price Range')
```

```python
sns.histplot(df['calculated_host_listings_count'], bins=30, kde=True, color='coral')
plt.title('Distribution of Host Listing Counts')
plt.xlabel('Listings per Host')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.violinplot(data=df[df['price'] < df['price'].quantile(0.95)],
               x='neighbourhood_group', y='price', palette='Accent')
plt.title('Price Range by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.show()
```

```python
room_dist = pd.crosstab(df['neighbourhood_group'], df['room_type'])
room_dist.plot(kind='bar', stacked=True, colormap='Set3', figsize=(10,6))
plt.title('Room Type Distribution by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Count')
plt.legend(title='Room Type')
plt.show()
```

```python
avg_price_neigh = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(20)
sns.barplot(x=avg_price_neigh.values, y=avg_price_neigh.index, palette='magma')
plt.title('Top 20 Neighbourhoods by Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.show()


```

```python
top_neigh = df['neighbourhood'].value_counts().head(5).index
df[df['neighbourhood'].isin(top_neigh)].groupby(['neighbourhood', 'room_type']).size().unstack().plot(kind='bar', stacked=True, title='Room Types in Top 5 Neighbourhoods')
```

```python
df['availability_bucket'] = pd.cut(df['availability_365'], bins=[0, 90, 180, 270, 365],
                                    labels=['090', '91180', '181270', '271365'])

sns.countplot(data=df, x='availability_bucket', palette='Set1')
plt.title('Listing Count by Availability Bucket')
plt.xlabel('Availability (Days)')
plt.ylabel('Number of Listings')
plt.show()
```

```python
room_type_counts = df['room_type'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Room Types')
plt.axis('equal')
plt.show()
```

```python
df.groupby('room_type')['price'].mean().sort_values().plot(kind='bar', title='Average Price by Room Type', color='skyblue')
```

```python
df.groupby('neighbourhood_group')['price'].mean().sort_values().plot(kind='bar', color='tomato', title='Average Price per Neighbourhood Group')
```

 Conclusion:
In this project, we performed a complete Exploratory Data Analysis (EDA) on the Airbnb NYC 2019 dataset. Our goal was to understand how listings are distributed across New York City, what types of properties are available, how prices vary, and which areas and room types are more popular.

We started by cleaning the data  handling missing values in important columns like name, last_review, reviews_per_month, and calculated_host_listings_count. After cleaning, we used several visualizations to explore patterns and trends.

Here are the key insights we discovered:

Manhattan and Brooklyn are the most common neighbourhoods for listings, with Manhattan having the highest average price.

The majority of listings are entire homes/apartments, but private rooms are also widely available and tend to get more frequent reviews.

Price distributions are skewed, with many low-cost listings and a few very expensive ones.

Hosts with a single listing make up a large portion of the platform, but there are also some hosts managing dozens or even hundreds of listings.

Reviews per month give us an idea about how popular and frequently booked a listing is.

We also created creative and informative visualizations like histograms, pie charts, bar plots, scatter plots, and correlation heatmaps to support our analysis.

This project helped us understand Airbnb's structure in New York City and can be useful for both guests (to find better deals) and hosts (to improve their listings). It also shows how data analysis can turn raw data into meaningful insights.
